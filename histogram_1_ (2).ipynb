```python
# Histogram Equalization

Url = "https://th.bing.com/th/id/R.2fc083ff8c2eb5bbbe8394db4fd6bc88?rik=XH10drlP0zIDEw&riu=http%3a%2f%2fupload.wikimedia.org%2fwikipedia%2fcommons%2f9%2f9a%2fGymnaster_savatieri1_flower.jpg&ehk=6FsNrc8LXS3O7uSJ8P4Cs%2fIJ%2fwK3Z510eFqK1%2fiHkmU%3d&risl=&pid=ImgRaw&r=0"
# Step 1: Download the image from the internet
response = requests.get(Url)
response.raise_for_status()  # Ensure the request was successful

# Step 2: Load the image into memory using BytesIO
image_bytes = BytesIO(response.content)
image = io.imread(image_bytes)

# Convert to grayscale
grayscale_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_img = cv2.equalizeHist(grayscale_img)

# Display the original, grayscale, and equalized images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(grayscale_img, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(1, 3, 3)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')

# Show the histogram of the equalized image
plt.figure()
plt.hist(equalized_img.ravel(), 256, [0, 256])
plt.show()
```

```python
import requests
from io import BytesIO
import cv2
from skimage import io
import matplotlib.pyplot as plt

# Step 1: Download the image from the internet
url = "https://th.bing.com/th/id/R.2fc083ff8c2eb5bbbe8394db4fd6bc88?rik=XH10drlP0zIDEw&riu=http%3a%2f%2fupload.wikimedia.org%2fwikipedia%2fcommons%2f9%2f9a%2fGymnaster_savatieri1_flower.jpg&ehk=6FsNrc8LXS3O7uSJ8P4Cs%2fIJ%2fwK3Z510eFqK1%2fiHkmU%3d&risl=&pid=ImgRaw&r=0"
response = requests.get(url)
response.raise_for_status()  # Ensure the request was successful

# Step 2: Load the image into memory using BytesIO
image_bytes = BytesIO(response.content)
image = io.imread(image_bytes)

# Convert to grayscale
grayscale_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_img = cv2.equalizeHist(grayscale_img)

# Display the original, grayscale, and equalized images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(grayscale_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()

# Show the histogram of the equalized image
plt.figure()
plt.hist(equalized_img.ravel(), 256, [0, 256])
plt.title('Histogram of Equalized Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```
