```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import iradon, radon

# Create a phantom image
def create_phantom():
    from skimage.data import shepp_logan_phantom
    phantom = shepp_logan_phantom()
    return phantom

# Simulate projection data
def simulate_projections(image, angles):
    sinogram = radon(image, theta=angles, circle=True)
    return sinogram

# Reconstruct image using FBP
def reconstruct_fbp(sinogram, angles):
    reconstruction_fbp = iradon(sinogram, theta=angles, filter_name='ramp')
    return reconstruction_fbp

# Main function
def main():
    image = create_phantom()
    angles = np.linspace(0., 180., max(image.shape), endpoint=False)
    sinogram = simulate_projections(image, angles)
    reconstruction_fbp = reconstruct_fbp(sinogram, angles)

    plt.figure(figsize=(8, 4.5))
    plt.subplot(131), plt.imshow(image, cmap='gray')
    plt.title('Original Phantom')
    plt.subplot(132), plt.imshow(sinogram, cmap='gray')
    plt.title('Sinogram')
    plt.subplot(133), plt.imshow(reconstruction_fbp, cmap='gray')
    plt.title('Reconstruction FBP')
    plt.show()

if __name__ == '__main__':
    main()


```
